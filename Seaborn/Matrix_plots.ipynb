```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tc = tips.select_dtypes(include=[np.number]).corr()
```

```python
sns.heatmap(tc,annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='black',linewidths=0.7)
```

```python
sns.clustermap(fp, cmap='coolwarm', linecolor='black',linewidths=0.5, standard_scale=1)
```
