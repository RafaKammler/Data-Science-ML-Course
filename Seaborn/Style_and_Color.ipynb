```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data = tips)
sns.despine(left=True)
```

```python
sns.set_context(context='notebook')
sns.countplot(x='sex', data = tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```
